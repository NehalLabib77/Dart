# Dart Foundations Assignment
This assignment will help you practice the concepts of Dart foundations like Variables, Nullability, Operators, String Manipulation, and Control Flow.

### 1. Variables and Nullability
Create a nullable String variable and assign it a value. Print the length of the variable.
Use the null-aware operator to handle if the variable is null.


### 2. Operators
Perform some basic arithmetic operations. Calculate the total price after applying a discount.
Use the `??` operator to handle null values in the calculation.


### 3. String Manipulation
Use string interpolation to print a formatted message containing variables.
Also, show how to manipulate strings (concatenate, change case, etc.).


### 4. Nullability and Default Value
Create a nullable string variable and use the null-coalescing operator (`??`) to provide a default value when the variable is null.


### 5. Control Flow: if/else
Create a conditional check that gives a message based on the number of hours a student studied.
Print different messages for study hours greater than 8, between 4-8, and less than 4 hours.


### 6. Switch Statement
Using a switch-like control flow, give a response based on the user's mood.
Possible moods: 'happy', 'sad', 'neutral'. Print different responses for each mood.


### 7. Looping: for, while
Write a `for` loop to print the first 10 even numbers.


### 8. Nested Loops
Create a nested loop that prints a multiplication table (up to 5x5).
